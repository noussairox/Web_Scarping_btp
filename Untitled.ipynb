```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager

# Configurer Selenium avec le WebDriver pour Edge
options = webdriver.EdgeOptions()
options.add_argument("--headless")  # Pour exécuter le navigateur en mode headless
driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)

# URL de la page à scraper
url = "https://www.kerix.net/fr/annuaire-entreprise/pieces-adaptables-pieces-dusure-btp.html"

# Ouvrir la page
driver.get(url)
time.sleep(50)  # Attendre que la page se charge

# Fonction pour extraire les données des entreprises
def extract_data(driver):
    entreprises = []
    
    # Trouver tous les éléments contenant les entreprises
    entreprise_elements = driver.find_elements(By.CSS_SELECTOR, 'div.col-lg-12')
    print(f"Nombre d'entreprises trouvées : {len(entreprise_elements)}")
    
    for index, element in enumerate(entreprise_elements):
        try:
            nom_element = element.find_element(By.TAG_NAME, 'h2')
            nom = nom_element.text.strip() if nom_element else "N/A"
            print(f"Nom de l'entreprise {index + 1} : {nom}")
            
            voir_plus = element.find_element(By.LINK_TEXT, 'Voir plus')
            voir_plus.click()
            time.sleep(2)  # Attendre que la page des détails se charge
            print(f"Accès aux détails de l'entreprise {index + 1} réussi.")
            
            adresse_element = driver.find_element(By.CSS_SELECTOR, 'p.card-text')
            adresse = adresse_element.text.strip() if adresse_element else "N/A"
            print(f"Adresse de l'entreprise {index + 1} : {adresse}")
            
            telephone_elements = driver.find_elements(By.CSS_SELECTOR, 'p.label-tel')
            telephones = [tel.text.strip() for tel in telephone_elements]
            telephone = ", ".join(telephones) if telephones else "N/A"
            print(f"Téléphone(s) de l'entreprise {index + 1} : {telephone}")
            
            entreprises.append([nom, adresse, telephone])
            
            driver.back()
            time.sleep(2)  # Attendre que la page principale se recharge
        except Exception as e:
            print(f"Erreur lors de l'extraction des données pour l'entreprise {index + 1} : {e}")
            continue
    
    return entreprises

# Extraire les données
data = extract_data(driver)
print("Extraction des données terminée.")

# Sauvegarder les données dans un fichier CSV
def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Nom", "Adresse", "Téléphone"])
        writer.writerows(data)

# Sauvegarder les données extraites
save_to_csv(data, 'fournisseurs.csv')
print("Les données ont été sauvegardées dans fournisseurs.csv")

# Fermer le navigateur
driver.quit()
```

```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager

# Configurer Selenium avec le WebDriver pour Edge
options = webdriver.EdgeOptions()
# Ne pas utiliser le mode headless pour contourner Cloudflare
# options.add_argument("--headless")

# Initialiser le WebDriver
driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)

# URL de la page à scraper
url = "https://www.kerix.net/fr/annuaire-entreprise/pieces-adaptables-pieces-dusure-btp.html"

# Ouvrir la page
driver.get(url)
print("Page chargée")
time.sleep(15)  # Attendre que la page se charge complètement

# Fonction pour extraire les données des entreprises
def extract_data(driver):
    entreprises = []
    
    # Trouver tous les éléments contenant les entreprises
    entreprise_elements = driver.find_elements(By.CSS_SELECTOR, 'div.col-lg-12')
    print(f"Nombre d'entreprises trouvées : {len(entreprise_elements)}")
    
    for index, element in enumerate(entreprise_elements):
        try:
            voir_plus = element.find_element(By.CSS_SELECTOR, 'a.btn-success')
            voir_plus.click()
            time.sleep(10)  # Attendre que la page des détails se charge
            print(f"Accès aux détails de l'entreprise {index + 1} réussi.")
            
            nom_element = driver.find_element(By.CSS_SELECTOR, 'h1.card-title')
            nom = nom_element.text.strip() if nom_element else "N/A"
            print(f"Nom de l'entreprise {index + 1} : {nom}")
            
            adresse_element = driver.find_element(By.CSS_SELECTOR, 'p.card-text')
            adresse = adresse_element.text.strip() if adresse_element else "N/A"
            print(f"Adresse de l'entreprise {index + 1} : {adresse}")
            
            # Cliquer sur le bouton "téléphone" pour afficher les numéros de téléphone
            try:
                telephone_button = driver.find_element(By.CSS_SELECTOR, 'a.telCli')
                telephone_button.click()
                time.sleep(2)  # Attendre que les numéros de téléphone soient affichés
                
                telephone_elements = driver.find_elements(By.CSS_SELECTOR, 'div#collapseExample p.label-tel')
                telephones = [tel.text.strip() for tel in telephone_elements]
                telephone = ", ".join(telephones) if telephones else "N/A"
                print(f"Téléphone(s) de l'entreprise {index + 1} : {telephone}")
            except Exception as e:
                print(f"Erreur lors du clic sur le bouton téléphone pour l'entreprise {index + 1} : {e}")
                telephone = "N/A"
            
            entreprises.append([nom, adresse, telephone])
            
            driver.back()
            time.sleep(5)  # Attendre que la page principale se recharge
        except Exception as e:
            print(f"Erreur lors de l'extraction des données pour l'entreprise {index + 1} : {e}")
            continue
    
    return entreprises

# Extraire les données
data = extract_data(driver)
print("Extraction des données terminée.")

# Sauvegarder les données dans un fichier CSV
def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Nom", "Adresse", "Téléphone"])
        writer.writerows(data)

# Sauvegarder les données extraites
save_to_csv(data, 'fournisseursbtp.csv')
print("Les données ont été sauvegardées dans fournisseurs.csv")

# Fermer le navigateur
driver.quit()
```

```python
import pandas as pd
from geopy.geocoders import OpenCage
from geopy.extra.rate_limiter import RateLimiter

opencage_api_key = '56795ee3547443c2be3e0140d6fc477c'

# Charger le fichier CSV
df = pd.read_csv('fournisseursbtp.csv')

# Initialiser le géocodeur OpenCage
geolocator = OpenCage(api_key=opencage_api_key)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Fonction pour obtenir la latitude et la longitude
def get_lat_lon(address):
    try:
        location = geocode(address)
        return location.latitude, location.longitude
    except:
        return None, None

# Appliquer la fonction à chaque adresse
df['Latitude'], df['Longitude'] = zip(*df['Adresse'].apply(get_lat_lon))

# Sauvegarder le fichier CSV enrichi
df.to_csv('fournisseurs_with_coords.csv', index=False)

# Afficher les premières lignes pour vérifier
df.head()
```
